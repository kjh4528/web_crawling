### MLP
#### 활성화 함수
- 시그모이드 : 0 - 1 사이의 확률을 구하는 공식
- RELU : 입력값이 0 보다 작으면 0, 0보다 크면 입력값
#### 손실 함수
- MSE : 예측(회귀)
- CrossEntropy : 분류


```python
# MSE
y_true = np.array([1.0, 0.0, 1.0])
y_pred = np.array([0.9, 0.2, 0.8])

mse = np.mean((y_true - y_pred)**2)
print('MSE = ', mse)
```

```python
# CrossEntropy (확률 분포의 차이)
y_true = np.array([1.0, 0.0, 1.0])
y_pred = np.array([0.9, 0.2, 0.8])

ce = -np.sum(y_true * np.log(y_pred + 1e-9))
print('ce =',ce)
```

```python
# 경사하강법 (w와 b 최적화 과정)
def gd(lr, steps=20):
    w = 5.0 # 시작점
    history = [w]
    for _ in range(steps):
        grad = 2*w # f(W) = w^2
        w = w - lr*grad
        history.append(w)
    return history

for lr in [0.05,0.3,1.1]:
    h = gd(lr)
    print(h)

# 빨리 학습시키려고 학습률을 크게 주면 오히려 오차가 커짐 
```

```python
# MLP : XOR
np.random.seed(0)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
w1 = np.random.randn(2, 4) * 0.5
b1 = np.zeros((1, 4))
w2 = np.random.randn(4, 1) * 0.5
b2 = np.zeros((1, 1))
w1, b1, w2, b2
```

```python
# 오차함수
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
h = sigmoid(X @ w1 + b1)
y_pred = (h @ w2 + b2)
print('초기 예측:',y_pred.ravel().round(3))
```
