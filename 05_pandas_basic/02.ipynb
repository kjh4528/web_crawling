```python
import pandas as pd
import numpy as np
```

```python
orders = pd.DataFrame({
    'order_id':[1,2,3,4,5],
    'store':['강남','잠실','강남','홍대','강남'],
    'amount':[10000, 20000, 15000, 30000, 25000]
})
```

```python
store_count = orders['store'].value_counts()
store_count
```

```python
# 매장별 주문 건수화 총금액 함께 집계
orders.groupby('store').agg(order_count=('order_id','count'), total_amount=('amount','sum')).reset_index()
```

```python
# 고객 데이터
customers = pd.DataFrame({
    'customer_id':[101,102,103],
    'customer_name':['앨리스','밥','캐롤']
})
```

```python
orders = pd.DataFrame({
    'order_id':[1,2,3,4,5],
    'store':['강남','잠실','강남','홍대','강남'],
    'amount':[10000, 20000, 15000, 30000, 25000],
    'customer_id':[101,102,104,np.nan,np.nan]
})
orders
```

```python
# merge
inner_join = pd.merge(orders, customers, on='customer_id', how='inner')
inner_join
```

```python
left_join = orders.merge(customers, on='customer_id', how='left')
left_join 
```
